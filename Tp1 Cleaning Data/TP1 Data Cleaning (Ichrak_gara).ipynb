## <span style="color:	purple "> 7/ Cleaning columns using the `.apply` function  </span>

### Ecrire une fonction qui permet de nettoyer la colonne des dates 

````python
unwanted_characters = ['[', ',', '-']
# Completer la fonction suivante
def clean_dates(item):
    dop= str(item.loc['Date of Publication'])
    
    if dop == 'nan' or dop[0] == '[':
        return np.NaN
    
    for character in unwanted_characters:
        if character in dop:
            character_index = dop.find(character)
            dop = dop[:character_index]
    
    return dop

df['Date of Publication'] = df.apply(clean_dates, axis = 1)
````

## <span style="color:	purple "> 8/ Observer la collone `Author`  </span>

### Completer la fonction `clean_author_names`qui permet de corriger les valeurs de cet colonne

````python
# complter la fonction suivante (  la place de pass)
def clean_author_names(author):
    
    author = str(author)
    
    if author == 'nan':
        return 'NaN'
    
    author = author.split(',')

    if len(author) == 1:
        name = filter(lambda x: x.isalpha(), author[0])
        return reduce(lambda x, y: x + y, name)
    
    last_name, first_name = author[0], author[1]

    first_name = first_name[:first_name.find('-')] if '-' in first_name else first_name
    
    if first_name.endswith(('.', '.|')):
        parts = first_name.split('.')
        
        
        if len(parts) > 1:
            occ_1 = first_name.find('.')
            occ_n = first_name.find('.', occ_1+1)
            first_name = first_name[:occ_1]
        else:
            first_name = first_name[:first_name.find('.')]
    
    last_name = last_name.capitalize()
    
    return f'{first_name} {last_name}'


df['Author'] = df['Author'].apply(clean_author_names)
````

## <span style="color:	purple "> 9/ Observer la colone `title`  </span>

### Completer la fonction `clean_title`qui permet de corriger les valeurs de cet colonne

````python
## compltrer la fonction suivante (  la place de pass)
def clean_title(title):
    
    if title == 'nan':
        return 'NaN'
    
    if title[0] == '[':
        title=title.replace('[','')
    if 'by' in title:
        parts=title.split('by')
        title=parts[0]
    elif 'By' in title:
        parts=title.split('By')
        title=parts[0]
    if '[' in title:
        parts=title.split('[')
        title=parts[0]

        
    title = title[:-2]
        
    title = list(map(str.capitalize, title.split()))
    return ' '.join(title)
    
df['Title'] = df['Title'].apply(clean_title)
df.head()
````

## <span style="color:	purple "> 2/ Crer un pandas data frame  partir d'un fichier txt </span>
On veut crer un DataFrame nomm `towns_df`  partir du fichier 'university_towns.txt'

Nous voyons que nous avons des noms d'tats priodiques suivis des villes universitaires de cet tat: StateA TownA1 TownA2 StateB TownB1 TownB2 .... Si nous regardons la faon dont les noms d'tats sont crits dans le fichier, nous verrons qu'ils ont tous la sous-chane [edit] en eux.

Nous pouvons tirer parti de ce modle en crant une liste de tuples (tat, ville) et en enveloppant cette liste dans un DataFrame:

````python
university_towns = []
with open('Datasets/university_towns.txt') as file:
    for line in file:
        if '[edit]' in line:
             # se rappeler de ce `state` jusqu' ce qu'on trouve le suivant
            state = line
        else:
             # Sinon, nous avons une ville; garder `state` 
            university_towns.append((state, line))

university_towns[:5]
````

````python
towns_df = pd.DataFrame(university_towns,
                        columns=['State', 'RegionName'])
towns_df.head()
````

## <span style="color:	purple "> 3/ Complter la fonction `clean_up` pour nettoyer les colonnes de `towns_df` </span>


````python
## compltrer la fonction suivante (  la place de pass)
def clean_up(item):
    if '(' in item:
        
        return item[:item.find('(') - 1]
    
    if '[' in item:
        return item[:item.find('[')]
    

towns_df =  towns_df.applymap(clean_up)
towns_df
````
